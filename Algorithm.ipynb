```python
# 9184

def w(a, b, c):
    global dp

    if a <= 0 or b <= 0 or c <= 0:
        return 1
    
    if a > 20 or b > 20 or c > 20:
        return w(20, 20, 20)

    if dp[a][b][c]:
        return dp[a][b][c]

    if a < b and b < c:
        dp[a][b][c] = w(a, b, c-1) + w(a, b-1, c-1) - w(a, b-1, c)
        return dp[a][b][c]
    else:
        dp[a][b][c] = w(a-1, b, c) + w(a-1, b-1, c) + w(a-1, b, c-1) - w(a-1, b-1, c-1)
        return dp[a][b][c]

dp = [[[0 for i in range(21)] for j in range(21)] for k in range(21)]
while True:
    a, b, c = map(int, input().split())
    if a == -1 and b == -1 and c == -1:
        break

    print('w({}, {}, {}) = {}'.format(a, b, c, w(a, b, c)))
```

```python
# 14889

N = int(input())

graph = []
for _ in range(N):
    graph.append(list(map(int, input().split())))

start_team = []
link_team = [i for i in range(N)]
min_size = 1e9
def solve(depth):
    global N, min_size

    if len(start_team) == N/2:
        start_score = 0
        link_score = 0
        for i in range(int(N/2)):
            for j in range(i+1, int(N/2)):
                start_score += graph[start_team[i]][start_team[j]] + graph[start_team[j]][start_team[i]]

        for i in range(int(N/2)):
            for j in range(i+1, int(N/2)):
                link_score += graph[link_team[i]][link_team[j]] + graph[link_team[j]][link_team[i]]

        min_size = min(min_size, abs(start_score - link_score))
        return

    for i in range(depth, N):
        start_team.append(i)
        link_team.remove(i)
        solve(i+1)
        start_team.pop()
        link_team.append(i)

solve(0)
print(min_size)
```

```python
# 2580

board = []
for _ in range(9):
    row = list(map(int, input().split()))
    board.append(row)

blank = []
for i in range(9):
    for j in range(9):
        if board[i][j] == 0:
            blank.append((i,j))

def checkRow(x, n):
    for j in range(9):
        if n == board[x][j]:
            return False
    
    return True
        

def checkCol(y, n):
    for i in range(9):
        if n == board[i][y]:
            return False
    
    return True

def checkRect(x, y, n):
    nx = x // 3 * 3
    ny = y // 3 * 3
    
    for i in range(3):
        for j in range(3):
            if board[nx+i][ny+j] == n:
                return False

    return True

def dfs(i):
    if len(blank) == i:
        for row in board:
            print(*row)
        exit(0)

    x = blank[i][0]
    y = blank[i][1]
    for n in range(1, 10):
        if checkRow(x, n) and checkCol(y, n) and checkRect(x, y, n):
            board[x][y] = n
            dfs(i+1)
            board[x][y] = 0
        
dfs(0)
```

```python
# 1018

def check(color, chess):
    cnt = 0
    for r in chess:
        for s in r:
            if s != color:
                cnt += 1
            
            if color == 'W':
                color = 'B'
            else:
                color = 'W'

        if color == 'W':
            color = 'B'
        else:
            color = 'W'
    
    return cnt
    

N, M = map(int, input().split())

board = []
for _ in range(N):
    board.append(input())

result = 9999
for i in range(N-8+1):
    for j in range(M-8+1):
        chess = []

        for row in board[i:i+8]:
            chess.append(row[j:j+8])

        cnt = min(check('W', chess), check('B', chess))
        result = min(result, cnt)

print(result)
```

```python
# 11729

N = int(input())

print(2 ** N - 1) # 최소 이동 횟수

def hanoi(n, from_pos, to_pos, aux_pos):
    
    if n == 1:
        print(from_pos, to_pos)
        return

    hanoi(n-1, from_pos, aux_pos, to_pos) # n-1개 보조 타워로 이동
    print(from_pos, to_pos) # 가장 큰 원판 목표 타워로 이동
    hanoi(n-1, aux_pos, to_pos, from_pos) # n-1개 목표 타워로 이동

hanoi(N, 1, 3, 2)
```

```python
# 2447

def printStar(n, rect):
    x = []
    if n == 3:
        return rect

    for i in rect:
        x.append(i * 3)
    
    for i in rect:
        x.append(i+' '*len(rect)+i)

    for i in rect:
        x.append(i * 3)

    return printStar(n//3, x)

N = int(input())

first = ['***', '* *', '***']
first = printStar(N, first)

for res in first:
    print(res)
```
